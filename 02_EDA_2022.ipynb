```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
print("Setup Complete")
```

```python
pass_age = db.fillna(0).sort_values(by=['Age'], ascending=True).reset_index()

pass_age
```

```python
age_surv = db.groupby('Age').sum()
age_surv
```

```python
plt.figure(figsize=(20,6))
plt.title("Survival of passengers according to age")

sns.barplot(x=age_surv.index, y=age_surv['Survived'])
```

```python
age_pass_all = db['Age'].value_counts().sort_index().to_frame()
age_pass_all
```

```python
plt.figure(figsize=(20,6))
plt.title("Age of passengers")

sns.lineplot(data=age_pass_all)
```
