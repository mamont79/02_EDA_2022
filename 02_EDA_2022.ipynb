```python
pass_age = db.fillna(0).sort_values(by=['Age'], ascending=True).reset_index()

pass_age
```

```python
age_surv = db.groupby('Age').sum().reset_index()
age_surv.drop(['PassengerId', 'Pclass'],axis=1)
```

```python
plt.figure(figsize=(100,10))
plt.title("Survival of passengers according to age")

sns.barplot(x=age_surv['Age'], y=age_surv['Survived'])
```

```python
По граффику можно обратить внимание на возраста от 23 до 30, и от 47 до 58. В этих диапазонах наибольшее количество выживших.
Причём в диапазоне 47-58 выживших больше
```

```python
age_pass_all = db['Age'].value_counts().sort_index().to_frame()
age_pass_all
```

```python
plt.figure(figsize=(20,6))
plt.title("Age of passengers")

sns.lineplot(data=age_pass_all)
```

```python
По общему количеству большую часть часть пассажиров  по возрасту попадают в диапазон 23-30 лет и 47-55
При сравнении с предыдущим графиком можно утверждать, что процентное соотношение выживших в диапазоне 23-30 лет
значительно ниже чем в остальных диапазонах.
```

```python
Линия тренда имеет положительное направление роста, 
возможно есть связь между количеством выживших и увеличением возраста.

Сгруппируем данные о выживших по возрасту и по полу 
```

```python
age_surv_sex = db.groupby(['Age', 'Sex']).sum().reset_index()
age_surv_sex
```

```python
sns.scatterplot(x=age_surv_sex['Age'], y=age_surv_sex['Survived'], hue=age_surv_sex['Sex'])
```

```python
sns.lmplot(x="Age", y="Survived", hue="Sex", data=age_surv_sex)
```

```python
При разделении на женщин и мужчин, можно заметить что женщины чаще выживают при кораблекрушении.
```

```python
plt.figure(figsize=(20,8))
sns.swarmplot(x=age_surv_sex['Sex']  , y=age_surv_sex['Survived'])
```
